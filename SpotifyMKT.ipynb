```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
df_spotify_albums = pd.read_csv(r"C:\Programa√ß√£o\MKT + Dados\spotify_albums.csv")
df_spotify_artists = pd.read_csv(r"C:\Programa√ß√£o\MKT + Dados\spotify_artists.csv")
df_spotify_tracks = pd.read_csv(r"C:\Programa√ß√£o\MKT + Dados\spotify_tracks.csv")


```

```python
pd.set_option('display.max_columns', None)  #mostra todas as colunas
pd.set_option('display.max_rows', 10)  #numero de linhas a mostrar
pd.reset_option('display.max_colwidth')
pd.reset_option('display.width')

print("üéµ Spotify Albums:")
display(df_spotify_albums.head())

print("\nüé§ Spotify Artists:")
display(df_spotify_artists.head())

print("\nüé∂ Spotify Tracks:")
display(df_spotify_tracks.head())
```

```python
print(df_spotify_albums.columns)

colunas_para_remover = ['available_markets', 'href', 'copyrights', 'external_urls', 'release_date_precision', 'type', 'Unnamed: 0', 'track_id', 'track_name_prev']

df_spotify_albums.rename(columns={'id': 'album_id'}, inplace=True)
df_spotify_albums.rename(columns={'name': 'album_name'}, inplace=True)
df_spotify_albums.rename(columns={'uri': 'album_uri'}, inplace=True)


df_spotify_albums = df_spotify_albums.drop(columns=colunas_para_remover, errors='ignore')
```

```python
print(df_spotify_artists.columns)
colunas_para_remover_artistas = ['Unnamed: 0','type', 'track_name_prev']
df_spotify_artists.rename(columns={'id': 'artists_id'}, inplace=True)
df_spotify_artists.rename(columns={'Name': 'artists_name'}, inplace=True)
df_spotify_artists = df_spotify_artists.drop(columns=colunas_para_remover_artistas, errors='ignore')


```

```python
print(df_spotify_tracks.columns)
colunas_para_remover_tracks = ['Unnamed: 0', 'artists_id', 'track_name_prev', 'acousticness', 'danceability','duration_ms', 'href', 'energy', 'instrumentalness', 'key', 'liveness', 'loudness', 'mode', 'popularity', 'speechiness', 'tempo', 'time_signature', 'valence']
df_spotify_tracks = df_spotify_tracks.drop(columns=colunas_para_remover_tracks, errors='ignore')
df_spotify_tracks.rename(columns={'id': 'tracks_id'}, inplace=True)
df_spotify_tracks.rename(columns={'name': 'track_name'}, inplace=True) 
df_spotify_tracks.rename(columns={'uri': 'tracks_uri'}, inplace=True)
```

```python
print("üéµ Spotify Albums:")
display(df_spotify_albums.head())

print("\nüé§ Spotify Artists:")
display(df_spotify_artists.head())

print("\nüé∂ Spotify Tracks:")
display(df_spotify_tracks.head())
```

```python
df_merged = pd.merge(df_spotify_albums, df_spotify_tracks, how='left', left_on='album_id', right_on='album_id')
df_final =pd.merge(df_merged, df_spotify_artists, how='left', left_on='artist_id', right_on='artists_id')
print ("üéµ Merged eeeeeeeeeeeeeeeeeeeeeeeeeeee")

pd.set_option('display.max_columns', None) #mostra todas as colunas

display(df_final.head())
```

```python
df_final.to_csv("df_final.csv", index=False)
print("Arquivo salvo como df_final.csv")
```
